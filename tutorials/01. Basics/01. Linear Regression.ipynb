## Prepare data

```python
import random
import numpy as np

xs = np.linspace(0, 3, 120)
ys = 5 * xs + 2 + random.random()
```

## Draw prepared data

```python
import matplotlib.pyplot as plt

plt.plot(xs, ys, '.')
plt.show()    
```

```python
import tensorflow as tf
from random import randint

def loss(y, pred):
    return tf.reduce_mean(tf.square(y - pred))

epochs = 100
learning_rate = 0.15

W = tf.Variable(20, dtype=np.float32)
b = tf.Variable(30, dtype=np.float32)

for count in range(epochs):
    with tf.GradientTape() as t:        
        pred = W * xs + b
        loss = tf.reduce_mean(tf.square(ys - pred))
        weight, bias = t.gradient(loss, [W, b])    
        
        print(f"Epoch {count}, W: {W.numpy()}, b: {b.numpy()}, loss: {loss}")
        W.assign_sub(learning_rate * weight)
        b.assign_sub(learning_rate * bias)
```

## Estimated Value

```python
print(f"W: {W.numpy()}, b: {b.numpy()}")

plt.plot(xs, ys, '.', color='blue')

minx = min(xs)
miny = b.numpy()
maxx = max(xs)
maxy = W.numpy() * maxx + b.numpy()

plt.plot([minx,maxx], [miny,maxy], color='red')

plt.show()    
```
